```python
df.rename(columns={'Company':'Company_Name','TypeName':'Type_Name', 'ScreenResolution':'Screen_Resolution'},inplace=True)
df.columns = df.columns.str.lower()
df
```

```python
df['inches'] = df['inches'].replace(['-', '?', '', "-'"], None)
df['inches'] = pd.to_numeric(df['inches'], errors='coerce')
df
```

```python
df['inches'].unique()
```

```python
df = df.dropna(subset=['inches'])
```

```python
df['inches'].isnull().sum()
```

```python
df.dtypes
```

```python
df['screen_resolution'].unique()
```

```python
df.isnull().sum()
```

```python
df['ram'] = df['ram'].astype(str).str.replace('GB', '').astype(int)
df
```

```python
df['memory'] = df['memory'].replace('?', None)

# TB to GB
df['memory'] = df['memory'].str.replace('TB', '000GB', regex=False)
df['memory_size'] = df['memory'].str.extract(r'(\d+)').astype(float)
df['memory_size'] = df['memory_size'].fillna(0).astype(int)
```

```python
df
```

```python
df['memory_size'].isna().sum()
```

```python
full = df.pop('memory_size')
df.insert(7, 'memory_size', full)
df
```

```python
df.isnull().sum()
```

```python
df['memory'] = df['memory'].fillna('Unknown')
df
```

```python
df.isna().sum()
```

```python
df['resolution'] = df['screen_resolution'].str.extract(r'(\d+x\d+)')
df[['x_res', 'y_res']] = df['resolution'].str.split('x', expand=True)

df['x_res'] = df['x_res'].astype(int)
df['y_res'] = df['y_res'].astype(int)
df
```

```python
res = df.pop('resolution')
x = df.pop('x_res')
y = df.pop('y_res')
df.insert(4, 'resolution', res)
df.insert(5, 'x_res', x)
df.insert(6, 'y_res', y)
df
```

```python
df['opsys'].unique()
```

```python
df['opsys'] = df['opsys'].replace({'Mac OS X': 'macOS', 'macOS': 'macOS', 'Windows 10': 'Windows','Windows 7': 'Windows','No OS': 'No OS','Linux': 'Linux','Chrome OS': 'Chrome OS'})
df
```

```python
df['weight'].unique()
```

```python
df['weight'] = df['weight'].replace('?', None)
```

```python
df['weight'] = df['weight'].astype(str).str.replace('kg', '')
df['weight'] = pd.to_numeric(df['weight'], errors='coerce')
df['weight'] = df['weight'].astype(float)
df
```

```python
df.dtypes
```

```python
df['price'].dtype
```

```python
df['price'] = df['price'].astype(int)
df
```

```python
df.isnull().sum()
```

```python
df['weight'] = df['weight'].fillna(df['weight'].median())
df
```

#UNIVARIATE ANALYSIS

```python
# Company distribution
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='company_name')
plt.xticks(rotation=45)
plt.title("Company Distribution")
plt.show()

# Type distribution
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='type_name')
plt.xticks(rotation=45)
plt.title("Laptop Type Distribution")
plt.show()

# Operating system distribution
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='opsys')
plt.xticks(rotation=45)
plt.title("Operating System Distribution")
plt.show()

# RAM distribution
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='ram')
plt.title("RAM Distribution")
plt.show()

# Inches distribution
plt.figure(figsize=(10,5))
sns.histplot(df['inches'], kde=True, bins=20)
plt.title("Screen Size Distribution (inches)")
plt.show()

# Weight distribution
plt.figure(figsize=(10,5))
sns.histplot(df['weight'], kde=True, bins=20)
plt.title("Weight Distribution (kg)")
plt.show()

# Price distribution
plt.figure(figsize=(10,5))
sns.histplot(df['price'], kde=True, bins=25)
plt.title("Price Distribution")
plt.show()
```

#BIVARIATE ANALYSIS

```python
# Price vs Company
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x="company_name", y="price")
plt.xticks(rotation=45)
plt.title("Price vs Company")
plt.show()

# Price vs Type
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x="type_name", y="price")
plt.xticks(rotation=45)
plt.title("Price vs Type")
plt.show()

# Price vs RAM
plt.figure(figsize=(10,6))
sns.barplot(data=df, x="ram", y="price", estimator='mean')
plt.title("Average Laptop Price by RAM")
plt.show()

# Price vs Memory Size
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="memory_size", y="price")
plt.title("Price vs Memory Size (GB)")
plt.show()

# Price vs Screen Resolution (PPI)
df['ppi'] = ((df['x_res']**2 + df['y_res']**2)**0.5 / df['inches'])
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="ppi", y="price")
plt.title("Price vs PPI")
plt.show()
```
